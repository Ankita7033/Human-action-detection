```python
# Import all necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import resample
from sklearn.preprocessing import LabelEncoder, RobustScaler
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectFromModel, SelectKBest, f_classif
from sklearn.pipeline import make_pipeline
from sklearn.manifold import TSNE
import time
import warnings
import joblib
from scipy.stats import randint, uniform
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv('mhealth_raw_data.csv')

# ========================
# 1. Data Cleaning
# ========================
print("Initial shape:", df.shape)

# Check for missing values
print("\nMissing values:\n", df.isnull().sum())

# Remove duplicates if any
df = df.drop_duplicates()
print("\nShape after removing duplicates:", df.shape)

# Check for constant features
constant_cols = [col for col in df.columns if df[col].nunique() == 1]
if constant_cols:
    print("\nRemoving constant features:", constant_cols)
    df = df.drop(columns=constant_cols)

# ========================
# 2. EDA (Exploratory Data Analysis)
# ========================
# Activity distribution
plt.figure(figsize=(14, 6))
activity_label = {
    0: "None", 1: "Standing still", 2: "Sitting relaxing", 
    3: "Lying down", 4: "Walking", 5: "Climbing stairs",
    6: "Waist bends", 7: "Arm elevation", 8: "Knees bending",
    9: "Cycling", 10: "Jogging", 11: "Running", 12: "Jump front & back"
}

df['Activity'] = df['Activity'].map(activity_label)
activity_counts = df['Activity'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=activity_counts.index, y=activity_counts.values, palette='viridis')
plt.title('Activity Distribution')
plt.xlabel('Activity')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Balance the dataset
min_samples = 2000  # Minimum samples per class
balanced_dfs = []
for activity in df['Activity'].unique():
    activity_df = df[df['Activity'] == activity]
    if len(activity_df) > min_samples:
        activity_df = activity_df.sample(min_samples, random_state=42)
    balanced_dfs.append(activity_df)

df = pd.concat(balanced_dfs)
print("\nShape after balancing:", df.shape)

# Visualize sensor data distributions
sensor_cols = [col for col in df.columns if col not in ['Activity', 'subject']]
plt.figure(figsize=(16, 20))
for i, col in enumerate(sensor_cols[:12], 1):  # Plot first 12 sensors
    plt.subplot(4, 3, i)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# ========================
# 3. Feature Engineering
# ========================
# Encode labels
le = LabelEncoder()
df['Activity_encoded'] = le.fit_transform(df['Activity'])

# Subject encoding (if needed)
df['subject_encoded'] = LabelEncoder().fit_transform(df['subject'])

# Prepare features and target
X = df.drop(['Activity', 'Activity_encoded', 'subject'], axis=1)
y = df['Activity_encoded']

# ========================
# 4. Data Preprocessing
# ========================
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y)

# Feature scaling
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Dimensionality reduction (optional)
pca = PCA(n_components=0.95, random_state=42)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
print(f"\nReduced from {X_train_scaled.shape[1]} to {X_train_pca.shape[1]} features with PCA")

# ========================
# 5. Model Training with Optimization
# ========================
results = {}

def evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    start = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start
    
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    
    test_acc = accuracy_score(y_test, y_pred)
    train_acc = accuracy_score(y_train, y_pred_train)
    f1 = f1_score(y_test, y_pred, average='weighted')
    
    results[model_name] = {
        'model': model,
        'train_accuracy': train_acc,
        'test_accuracy': test_acc,
        'f1_score': f1,
        'train_time': train_time,
        'report': classification_report(y_test, y_pred, target_names=list(activity_label.values()))
    }
    
    print(f"\n{model_name} Results:")
    print(f"Train Accuracy: {train_acc:.4f}")
    print(f"Test Accuracy: {test_acc:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"Training Time: {train_time:.2f}s")
    print("\nClassification Report:")
    print(results[model_name]['report'])
    
    # Plot confusion matrix
    plt.figure(figsize=(12, 10))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=activity_label.values(),
                yticklabels=activity_label.values())
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# 5.1 Logistic Regression with GridSearch
print("\nTraining Logistic Regression...")
lr_params = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['saga'],
    'max_iter': [1000]
}
lr = GridSearchCV(LogisticRegression(), lr_params, cv=3, n_jobs=-1, verbose=1)
evaluate_model(lr, X_train_scaled, y_train, X_test_scaled, y_test, "Logistic Regression")

# 5.2 KNN with RandomizedSearch
print("\nTraining KNN...")
knn_params = {
    'n_neighbors': randint(3, 15),
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree']
}
knn = RandomizedSearchCV(KNeighborsClassifier(), knn_params, n_iter=10, cv=3, n_jobs=-1, random_state=42)
evaluate_model(knn, X_train_scaled, y_train, X_test_scaled, y_test, "KNN")

# 5.3 SVM with Reduced Search Space
print("\nTraining SVM...")
svm_params = {
    'C': [0.1, 1, 10],
    'kernel': ['rbf', 'linear'],
    'gamma': ['scale', 'auto']
}
svm = GridSearchCV(SVC(probability=True), svm_params, cv=2, n_jobs=-1, verbose=1)  # Reduced cv for speed
evaluate_model(svm, X_train_pca, y_train, X_test_pca, y_test, "SVM (with PCA)")

# 5.4 Naive Bayes (No tuning needed)
print("\nTraining Naive Bayes...")
nb = GaussianNB()
evaluate_model(nb, X_train_scaled, y_train, X_test_scaled, y_test, "Naive Bayes")

# 5.5 Decision Tree with GridSearch
print("\nTraining Decision Tree...")
dt_params = {
    'max_depth': [5, 10, 15, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
dt = GridSearchCV(DecisionTreeClassifier(random_state=42), dt_params, cv=3, n_jobs=-1, verbose=1)
evaluate_model(dt, X_train_scaled, y_train, X_test_scaled, y_test, "Decision Tree")

# 5.6 Random Forest with RandomizedSearch
print("\nTraining Random Forest...")
rf_params = {
    'n_estimators': randint(50, 200),
    'max_depth': [5, 10, 15, 20, None],
    'min_samples_split': [2, 5, 10],
    'max_features': ['sqrt', 'log2']
}
rf = RandomizedSearchCV(RandomForestClassifier(random_state=42, n_jobs=-1), 
                       rf_params, n_iter=10, cv=2, n_jobs=-1, verbose=1)
evaluate_model(rf, X_train_scaled, y_train, X_test_scaled, y_test, "Random Forest")

# ========================
# 6. Results Comparison
# ========================
results_df = pd.DataFrame.from_dict(results, orient='index')
results_df = results_df[['train_accuracy', 'test_accuracy', 'f1_score', 'train_time']]
results_df = results_df.sort_values('test_accuracy', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=results_df.index, y=results_df['test_accuracy'], palette='viridis')
plt.title('Model Comparison - Test Accuracy')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.xticks(rotation=45)
plt.ylim(0, 1)
for i, v in enumerate(results_df['test_accuracy']):
    plt.text(i, v + 0.02, f"{v:.3f}", ha='center')
plt.tight_layout()
plt.show()

# Save best model
best_model_name = results_df.index[0]
best_model = results[best_model_name]['model']
joblib.dump(best_model, 'best_model.pkl')

print("\nBest Model:", best_model_name)
print(results[best_model_name]['report'])
```
